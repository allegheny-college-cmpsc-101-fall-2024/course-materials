# Simple Scope Example

```python
# TODO: Comment to explain what is happening

from typing import List
from typing import Tuple

def prime_test_exhaustive(x: int) -> Tuple[bool, List[int]]:
  """Determine if input is prime. Return tuple of True for prime and factors in a list."""
  smallest_divisor = None
  for guess in range(2, x, 1):
    if x % guess == 0:
      smallest_divisor = guess
      break
  if smallest_divisor is not None:
    answer = (False, [smallest_divisor])
    return answer
  else:
    answer = (True, [1, x])
    return answer


number = 97
result = prime_test_exhaustive(number)
print(result)
```
